### 데이터 확인

```python
import numpy as np
import pandas as pd

df = pd.read_csv('c:/3rd_project/data/Train/movie_data_231009_total.csv')
df
```

```python
df['genre'].value_counts()
```

```python
# 장르 라벨링

from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
label_encoder = LabelEncoder()

print(df['genre'].unique())

# 'genre' 열을 라벨링하여 새로운 'label_genre' 열 생성
df['genre'] = label_encoder.fit_transform(df['genre'])
print(df['genre'].unique())

df
```

### 언더샘플링

```python
RANDOM_STATE = 2023
```

```python
train_cols = df.columns[1:-1]

X = df[train_cols]
y = df['genre']

from imblearn.under_sampling import RandomUnderSampler

X_sample, y_sample = RandomUnderSampler(random_state=RANDOM_STATE).fit_resample(X, y)
X_samp = pd.DataFrame(data=X_sample,columns=train_cols )
y_samp = pd.DataFrame(data=y_sample,columns=['genre'])

data_samp=pd.concat([X_samp,y_samp],axis=1)

data_samp["genre"].value_counts()
```

```python
data_samp
```

```python
X = data_samp.iloc[:, 4:-1].values
y = data_samp.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
data_samp
```

### 데이터 셋 나누기

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=RANDOM_STATE)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
```

```python
score_lists = []
model_lists = ['LogisticRegression', 'Tree', 'RandomForest', 'KNN', 'SVM', 'CNN']
```

로지스틱 선형 회귀

```python
# 로지스틱 회귀 모델 생성
logistic_model = LogisticRegression()

# 모델 훈련
logistic_model.fit(X_train, y_train)

# 모델 평가
y_pred = logistic_model.predict(X_test)
accuracy = round(accuracy_score(y_test, y_pred),4)

score_lists.append(accuracy)
```

```python
import joblib

# 로지스틱 회귀 모델을 저장
model_filename = "c:/3rd_project/model/logistic_model.pkl"
joblib.dump(logistic_model, model_filename)
```

단일 결정 트리

```python
# 단일 결정 트리 모델 생성
tree_model = DecisionTreeClassifier()

# 모델 훈련
tree_model.fit(X_train, y_train)

# 모델 평가
y_pred = tree_model.predict(X_test)
accuracy = round(accuracy_score(y_test, y_pred),4)

score_lists.append(accuracy)
```

```python
import joblib

# 단일 결정 트리 모델을 저장
model_filename = "c:/3rd_project/model/tree_model.pkl"
joblib.dump(tree_model, model_filename)
```

Random Forest

```python
# 랜덤 포레스트 모델 생성
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# 모델 훈련
rf_classifier.fit(X_train, y_train)

# 모델 평가
y_pred = rf_classifier.predict(X_test)
accuracy = round(accuracy_score(y_test, y_pred),4)

score_lists.append(accuracy)
```

```python
import joblib

# 랜덤 포레스트 모델을 저장
model_filename = "c:/3rd_project/model/random_forest_model.pkl"
joblib.dump(rf_classifier, model_filename)
```

KNN

```python
# KNN 모델 생성
k = 29  # K 값 설정
knn_model = KNeighborsClassifier(n_neighbors=k)

# 모델 훈련
knn_model.fit(X_train, y_train)

# 모델 평가
y_pred = knn_model.predict(X_test)
accuracy = round(accuracy_score(y_test, y_pred),4)

score_lists.append(accuracy)
```

```python
import joblib

# KNN 모델을 저장
model_filename = "c:/3rd_project/model/knn_model.pkl"
joblib.dump(knn_model, model_filename)
```

SVM

```python
# SVM 모델 생성
svm_classifier = SVC(kernel='linear', C=1.0)

# 모델 훈련
svm_classifier.fit(X_train, y_train)

# 모델 평가
y_pred = svm_classifier.predict(X_test)
accuracy = round(accuracy_score(y_test, y_pred),4)

score_lists.append(accuracy)
```

```python
import joblib

# SVM 모델을 저장
model_filename = "c:/3rd_project/model/svm_classifier.pkl"
joblib.dump(svm_classifier, model_filename)
```

CNN

```python
import tensorflow as tf

# 모델 호출
cnn_model = tf.keras.models.load_model('c:/3rd_project/model/CNN_231009_a.h5')

# 모델 평가
score = cnn_model.evaluate(X_test, y_test, verbose=1)
score_lists.append(round(score[1],4))
```

```python
score_lists
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # 그래프 크기 설정
bars = plt.bar(model_lists, score_lists, color='skyblue')  # 막대 그래프 생성

plt.xlabel('Model')  # x축 레이블 설정
plt.ylabel('Accuracy')  # y축 레이블 설정
plt.title('Model Accuracy')  # 그래프 제목 설정

plt.ylim(0, 1)  # y축 범위 설정 (0부터 1까지)

# 각 막대 그래프 중앙에 값을 표시
for bar, score in zip(bars, score_lists):
    plt.text(bar.get_x() + bar.get_width() / 2, score, f'{score:.4f}', ha='center', va='bottom')

plt.show()  # 그래프 표시
```

```python
print(f"최적 모델 : {model_lists[score_lists.index(max(score_lists))]}")
print(f"최고 정확도 : {score_lists[score_lists.index(max(score_lists))]}")
```

```python
# 새로운 데이터에 대한 예측
labels = {0: 'etc',
          1: 'horror',
          2: 'drama',
          3: 'ani',
          4: 'action',
          5: 'comedy'
          }
num = [885, 3925, 1732, 409, 486, 551]

for i in num:
    real_genre = df.iloc[i][-1]
    new_data = df.iloc[i][5:-1].values  # Pandas Series를 NumPy 배열로 변환
    new_data = new_data.reshape(1, -1)  # 2D 배열로 변환

    LR_predicted_genre = logistic_model.predict(new_data)
    Tree_predicted_genre = tree_model.predict(new_data)
    RF_predicted_genre = rf_classifier.predict(new_data)
    KNN_predicted_genre = knn_model.predict(new_data)
    SVM_predicted_genre = svm_classifier.predict(new_data)
    CNN_predicted_genre = cnn_model.predict(new_data.astype('float32'))

    print(f'실제 장르: {labels[real_genre]}')
    print(f'LR 모델-예측 장르: {labels[LR_predicted_genre[0]]}')
    print(f'Tree 모델-예측 장르: {labels[Tree_predicted_genre[0]]}')
    print(f'RF 포레스트 모델-예측 장르: {labels[RF_predicted_genre[0]]}')
    print(f'KNN 모델-예측 장르: {labels[KNN_predicted_genre[0]]}')
    print(f'SVM 모델-예측 장르: {labels[SVM_predicted_genre[0]]}')
    print(f'CNN 모델-예측 장르: {labels[np.argmax(CNN_predicted_genre)]}\n')
```
