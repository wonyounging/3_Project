```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('c:/3rd_project/data/Train/movie_data_231012_total.csv')
df
```

```python
act_genre = df[df['genre'] == '액션']
act_avg_freq = act_genre[['act_freq', 'ani_freq', 'com_freq', 'dra_freq', 'hor_freq']].mean()

ani_genre = df[df['genre'] == '애니메이션']
ani_avg_freq = ani_genre[['act_freq', 'ani_freq', 'com_freq', 'dra_freq', 'hor_freq']].mean()

com_genre = df[df['genre'] == '코미디']
com_avg_freq = com_genre[['act_freq', 'ani_freq', 'com_freq', 'dra_freq', 'hor_freq']].mean()

dra_genre = df[df['genre'] == '드라마']
dra_avg_freq = dra_genre[['act_freq', 'ani_freq', 'com_freq', 'dra_freq', 'hor_freq']].mean()

hor_genre = df[df['genre'] == '공포(호러)']
hor_avg_freq = hor_genre[['act_freq', 'ani_freq', 'com_freq', 'dra_freq', 'hor_freq']].mean()

etc_genre = df[df['genre'] == '-']
etc_avg_freq = etc_genre[['act_freq', 'ani_freq', 'com_freq', 'dra_freq', 'hor_freq']].mean()
```

```python
# 열 이름
columns = act_avg_freq.index

# 평균 값
act_freq_values = act_avg_freq.values
ani_freq_values = ani_avg_freq.values
com_freq_values = com_avg_freq.values
dra_freq_values = dra_avg_freq.values
hor_freq_values = hor_avg_freq.values
etc_freq_values = etc_avg_freq.values

width = 0.15  # 각 막대의 너비

plt.figure(figsize=(12, 6))

x = range(len(columns))  # X축 값

plt.bar(x, act_freq_values, width, label='act', color='#4169E1')
plt.bar([i + width for i in x], ani_freq_values, width, label='ani', color='#FF6347')
plt.bar([i + 2 * width for i in x], com_freq_values, width, label='com', color='#FFD700')
plt.bar([i + 3 * width for i in x], dra_freq_values, width, label='dra', color='#90EE90')
plt.bar([i + 4 * width for i in x], hor_freq_values, width, label='hor', color='#DDA0DD')
plt.bar([i + 5 * width for i in x], etc_freq_values, width, label='etc', color='#FF8C00')

plt.title('Genre Frequency')
plt.xlabel('Genre Frequency')
plt.ylabel('Avg Frequency')
plt.xticks([i + 2.5 * width for i in x], columns)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
