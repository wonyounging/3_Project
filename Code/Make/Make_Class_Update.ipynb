```python
import os

Gun_train_path = 'c:/3rd_project/data/Custom/Weapons Segmentation Dataset.v7i.yolov8/train/labels'
Gun_valid_path = 'c:/3rd_project/data/Custom/Weapons Segmentation Dataset.v7i.yolov8/valid/labels'
Gun_test_path = 'c:/3rd_project/data/Custom/Weapons Segmentation Dataset.v7i.yolov8/test/labels'
```

train label     0 => 80

```python
for filename in os.listdir(Gun_train_path):
    if filename.endswith(".txt"):  # .txt 확장자를 가진 파일만 처리하도록 조건을 설정합니다.
        file_path = os.path.join(Gun_train_path, filename)  # 파일의 전체 경로 생성

        # 파일 내용을 읽어옵니다.
        with open(file_path, 'r') as file:
            content = file.read()

        # 클래스 번호가 0으로 시작하는 경우 0을 8로 변경합니다.
        if content.startswith('0'):
            content = '80' + content[1:]

        # 수정된 내용을 파일에 다시 씁니다.
        with open(file_path, 'w') as file:
            file.write(content)

        print("클래스 번호 수정이 완료되었습니다.")
```
