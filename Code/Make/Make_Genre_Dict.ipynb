```python
import pandas as pd
```

```python
df = pd.read_excel("C:/3rd_project/data/Movie/[메인장르_최종] 영화포스터(2011.01-2023.07).xlsx", engine="openpyxl")
```

```python
df
```

```python
df = df[['메인장르', '줄거리']]
df
```

```python
dra_df = df[df['메인장르'] == ('드라마')]
com_df = df[df['메인장르'] == ('코미디')]
act_df = df[df['메인장르'] == ('액션')]
ani_df = df[df['메인장르'] == ('애니메이션')]
hor_df = df[df['메인장르'] == ("공포(호러)")]

dra_story_list = [] #
com_story_list = [] #
act_story_list = []
ani_story_list = []
hor_story_list = []

dra_nouns = [] #
com_nouns = [] #
act_nouns = []
ani_nouns = []
hor_nouns = []
```

```python
# 드라마

from tqdm import tqdm
from konlpy.tag import Kkma
import re

kkma = Kkma()

for story in dra_df['줄거리']:
    clean_review = re.sub(r'[^가-힣a-zA-Z\s]', '', story)
    dra_story_list.append(clean_review)

for story_nouns in tqdm(dra_story_list):
    dra_nouns.extend(kkma.nouns(story_nouns))
        
print(len(dra_nouns)) 
dra_nouns
```

```python
# 액션

from tqdm import tqdm
from konlpy.tag import Kkma
import re

kkma = Kkma()

for story in act_df['줄거리']:
    clean_review = re.sub(r'[^가-힣a-zA-Z\s]', '', story)
    act_story_list.append(clean_review)

for story_nouns in tqdm(act_story_list):
    act_nouns.extend(kkma.nouns(story_nouns))
        
print(len(act_nouns))
act_nouns
```

```python
# 코미디

from tqdm import tqdm
from konlpy.tag import Kkma
import re

kkma = Kkma()

for story in com_df['줄거리']:
    clean_review = re.sub(r'[^가-힣a-zA-Z\s]', '', story)
    com_story_list.append(clean_review)

for story_nouns in tqdm(com_story_list):
    com_nouns.extend(kkma.nouns(story_nouns))
        
print(len(com_nouns))
com_nouns
```

```python
# 애니메이션

from tqdm import tqdm
from konlpy.tag import Kkma
import re

kkma = Kkma()

for story in ani_df['줄거리']:
    clean_review = re.sub(r'[^가-힣a-zA-Z\s]', '', story)
    ani_story_list.append(clean_review)

for story_nouns in tqdm(ani_story_list):
    ani_nouns.extend(kkma.nouns(story_nouns))
        
print(len(ani_nouns))
ani_nouns
```

```python
# 공포

from tqdm import tqdm
from konlpy.tag import Kkma
import re

kkma = Kkma()

for story in hor_df['줄거리']:
    clean_review = re.sub(r'[^가-힣a-zA-Z\s]', '', story)
    hor_story_list.append(clean_review)

for story_nouns in tqdm(hor_story_list):
    hor_nouns.extend(kkma.nouns(story_nouns))

print(len(hor_nouns))
hor_nouns
```

```python
print(len(dra_nouns))
print(len(act_nouns))
print(len(com_nouns))
print(len(ani_nouns))
print(len(hor_nouns))
```

```python
dra_nouns = list(set(dra_nouns))
act_nouns = list(set(act_nouns))
com_nouns = list(set(com_nouns))
ani_nouns = list(set(ani_nouns))
hor_nouns = list(set(hor_nouns))
```

```python
# 텍스트 파일로 저장할 파일 경로
file_path = 'C:/3rd_project/data/hor_nouns.txt'

# 리스트를 텍스트 파일에 쓰기
with open(file_path, 'w', encoding='utf-8') as file:
    for noun in hor_nouns:
        if len(noun) > 1:
            file.write(noun + '\n')
```
