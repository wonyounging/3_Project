### 언더샘플링

```python
RANDOM_STATE = 2023
```

```python
train_cols = df.columns[1:-1]

X = df[train_cols]
y = df['genre']

from imblearn.under_sampling import RandomUnderSampler

X_sample, y_sample = RandomUnderSampler(random_state=RANDOM_STATE).fit_resample(X, y)
X_samp = pd.DataFrame(data=X_sample,columns=train_cols )
y_samp = pd.DataFrame(data=y_sample,columns=['genre'])

data_samp=pd.concat([X_samp,y_samp],axis=1)

data_samp["genre"].value_counts()
```

```python
data_samp
```

```python
X = data_samp.iloc[:, :-1].values
y = data_samp.iloc[:, -1].values
```

```python
#다양한 알고리즘별 정확도를 저장할 리스트

scores=[]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

#의사결정나무

from sklearn.tree import DecisionTreeClassifier

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', DecisionTreeClassifier(random_state=0)),
])

params = {
     'clf__max_depth': list(range(3,6))
}

grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, y)

print(grid.best_params_)
print(grid.best_score_)
```

```python
scores.append(grid.best_score_)
scores
```

```python
tree=grid.best_estimator_['clf']
tree
```

```python
#랜덤포레스트

from sklearn.ensemble import RandomForestClassifier

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', RandomForestClassifier(random_state=0)),
])

params = {
        'clf__max_depth': list(range(3,6)),
        'clf__n_estimators': list(range(10,110,10)),
}

grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, y)

print(grid.best_params_)
print(grid.best_score_)
```

```python
scores.append(grid.best_score_)
scores
```

```python
#knn

from sklearn.neighbors import KNeighborsClassifier

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', KNeighborsClassifier()),
])

params = {
       'clf__n_neighbors': list(range(1,11)),
}

grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, y)

print(grid.best_params_)
print(grid.best_score_)
```

```python
scores.append(grid.best_score_)
scores
```

```python
#인공신경망

from sklearn.neural_network import MLPClassifier

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', MLPClassifier(random_state=0, max_iter=5000)),
])

params = {
        'clf__hidden_layer_sizes': [(50,), (100,), (50,50), (100,100)],
        'clf__activation': ['tanh', 'relu'],
        'clf__alpha': [0.001, 0.01, 0.1]
}

grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, y)

print(grid.best_params_)
print(grid.best_score_)
```

```python
scores.append(grid.best_score_)
scores
```

```python
#svm

from sklearn.svm import SVC

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', SVC(random_state=0)),
])

params = {
        'clf__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
        'clf__gamma': [0.0001, 0.001, 0.01, 0.1],
}

grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, y)

print(grid.best_params_)
print(grid.best_score_)
```

```python
scores.append(grid.best_score_)
scores
```

```python
from scikeras.wrappers import KerasClassifier
from keras import models
from keras import layers

def create_network(optimizer="rmsprop"):
    network = models.Sequential()
    network.add(layers.Dense(units=16,
                         activation="relu",
                         input_shape=(X.shape[1],)))
    network.add(layers.Dense(units=16, activation="relu"))
    network.add(layers.Dense(units=1, activation="sigmoid"))    
    network.compile(loss='binary_crossentropy',
                    optimizer=optimizer, # 옵티마이저
                    metrics=["accuracy"]) # 성능 지표    

    return network
```

```python
neural_network = KerasClassifier(model=create_network, verbose=0)

# 하이퍼파라미터 탐색 영역을 정의

pipe=Pipeline([
    ('scaler',StandardScaler()),
    ('clf',neural_network)
])

params={
    'clf__epochs': [50, 100],
    'clf__batch_size': [32,64,128],
    'clf__optimizer': ["rmsprop", "adam"],
}

grid=GridSearchCV(pipe,params,scoring='accuracy',cv=3)

grid_result = grid.fit(X, y)
```

```python
grid_result.best_params_
```

```python
scores.append(grid.best_score_)
scores
```
