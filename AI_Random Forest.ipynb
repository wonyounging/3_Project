```python
# 장르 라벨링

from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
label_encoder = LabelEncoder()

print(df['genre'].unique())

# 'genre' 열을 라벨링하여 새로운 'label_genre' 열 생성
df['genre'] = label_encoder.fit_transform(df['genre'])
print(df['genre'].unique())

df
```

### 언더샘플링

```python
RANDOM_STATE = 2023
```

```python
train_cols = df.columns[1:-1]

X = df[train_cols]
y = df['genre']

from imblearn.under_sampling import RandomUnderSampler

X_sample, y_sample = RandomUnderSampler(random_state=RANDOM_STATE).fit_resample(X, y)
X_samp = pd.DataFrame(data=X_sample,columns=train_cols )
y_samp = pd.DataFrame(data=y_sample,columns=['genre'])

data_samp=pd.concat([X_samp,y_samp],axis=1)

data_samp["genre"].value_counts()
```

```python
data_samp
```

```python
X = data_samp.iloc[:, :-1].values
y = data_samp.iloc[:, -1].values
```

### 데이터 셋 나누기

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=RANDOM_STATE)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 랜덤 포레스트 모델 생성
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# 모델 훈련
rf_classifier.fit(X_train, y_train)
```

```python
# 모델 평가
y_pred = rf_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {round(accuracy,4)}')
```

```python
# 새로운 데이터에 대한 예측
num = [885, 3925, 1732, 409, 486, 551]

for i in num:

    real_genre = df.iloc[i][-1]
    new_data = df.iloc[i][1:-1].values  # Pandas Series를 NumPy 배열로 변환
    new_data = new_data.reshape(1, -1)  # 2D 배열로 변환
    predicted_genre = rf_classifier.predict(new_data)

    print(f'실제 장르: {real_genre}')
    print(f'예측 장르: {predicted_genre}\n')
```
